## Total unique search
Represent users interest to particular topic

```python
# The slices will be ordered and plotted counter-clockwise.
labels = df11.sort_values(by='Total Unique Searches', axis=0, ascending=False)['Search Term']
fracs = df11.sort_values(by='Total Unique Searches', axis=0, ascending=False)['Total Unique Searches']

plt.pie(fracs, labels=labels,
                autopct='%1.1f%%', shadow=True, startangle= 90)
                # The default startangle is 0, which would start
                # the Frogs slice on the x-axis.  With startangle=90,
                # everything is rotated counter-clockwise by 90 degrees,
                # so the plotting starts on the positive y-axis.

plt.title('Total Unique Searches')

plt.show()
```

## Average search depth
represents how far was result from original search.

**Example insigth - people were interested in galaxy note 5 at most, but they at average visited 3 pages to get to it after search, better to decrease this distance.**

```python
#setting size of the plots
fig_size = plt.rcParams["figure.figsize"]
fig_size[0] = 8
fig_size[1] = 8
plt.rcParams["figure.figsize"] = fig_size

n_groups = 10

means_men = list(df11.sort_values(by='Average Search Depth', axis=0, ascending=True)['Average Search Depth'])
titles = list(df11.sort_values(by='Average Search Depth', axis=0, ascending=True)['Search Term'])

fig, ax = plt.subplots()

index = np.arange(n_groups)
bar_width = 0.5


rects1 = plt.bar(index, means_men, bar_width,
                 color='c', alpha = 0.8)

plt.xlabel('Search term')
plt.ylabel('Number of pages visited after search term')
plt.title('Average Search Depth')
plt.xticks(index + bar_width, titles)

ax.grid(True)

for tick in ax.get_xticklabels():
    tick.set_rotation(45)

plt.legend()


plt.tight_layout()
plt.show();
```

## Searches with exit

The number of exits from site that occurred following a result from search.

**Example insigth - searches with more than 2 terms have high % of exit. It means that search engine possibly generates not relevant results when query consists from more than 2 terms.**

```python

n_groups = 10

means_men = list(df11.sort_values(by='% Search Exits', axis=0, ascending=True)['% Search Exits'])
titles = list(df11.sort_values(by='% Search Exits', axis=0, ascending=True)['Search Term'])

fig, ax = plt.subplots()

index = np.arange(n_groups)
bar_width = 0.5


rects1 = plt.bar(index, means_men, bar_width,
                 color='y', alpha = 0.8)

plt.xlabel('Search term')
plt.ylabel('% of exit')
plt.title('Searches with exit')
plt.xticks(index + bar_width, titles)

ax.grid(True)

for tick in ax.get_xticklabels():
    tick.set_rotation(45)

plt.legend()


plt.tight_layout()
plt.show();
```

```python
def autolabel(rects):
    # attach some text labels
    for rect in rects:
        height = rect.get_height()
        ax.text(rect.get_x() + rect.get_width()/2., 1.05*height,
                '%d' % int(height),
                ha='center', va='bottom')
```
